```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data={
    "Date":pd.to_datetime(["2022-01-01","2022-02-01","2022-03-01","2022-04-01","2022-05-01","2022-06-01","2022-07-01","2022-08-01","2022-09-01","2023-01-01","2023-02-01","2023-03-01","2023-04-01","2023-05-01","2023-06-01"]),
    "Value":[10,15,20,25,30,42,48,52,55,58,60,75,82,90,105]
}
```

```python
df=pd.DataFrame(data)
df.set_index("Date",inplace=True)
plt.figure(figsize=(10,6))

plt.plot(df.index,df["Value"],marker='X')
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Time Series with trend")
plt.grid(True)
plt.show()
```

```python
plt.plot(df.index,df["Value"],marker='o')
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Time Series with trend")
plt.grid(True)
plt.show()
```
