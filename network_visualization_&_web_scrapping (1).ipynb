```python
import networkx as nx
import matplotlib.pyplot as plt
G=nx.Graph()
G.add_nodes_from([1,2,3,4,5])
G.add_edges_from([(1,2),(1,3),(2,3),(3,4),(5,1)])
pos=nx.spring_layout(G)
nx.draw(G,pos,with_labels=True,node_color='skyblue',node_size=1400,edge_color='black',linewidths=1,font_size=15)
plt.show()
```

```python
def getdata(url):
    r = requests.get(url)
    return r.text
```

```python
htmldata = getdata("https://www.cars24.com/blog/10-most-expensive-cars-in-the-world/")
soup = BeautifulSoup(htmldata, 'html.parser')
for item in soup.find_all('img'):
    print(item)
```
